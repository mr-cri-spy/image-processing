```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread("/content/image1.tiff")
img2 = cv2.imread("/content/image2.tiff")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
images = [img1,img2]
for i in range(0,len(images)):
  plt.subplot(1,2,i+1)
  plt.imshow(images[i])
plt.show()
```

```python
## Weighted Addition of Images
```

```python
img1 = cv2.imread("/content/image1.tiff")
img2 = cv2.imread("/content/image2.tiff")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
output1 = cv2.addWeighted(img1,0.7,img2,0.3,0)
output2 = cv2.addWeighted(img1,0.7,img2,0.3,50)
output3 = cv2.addWeighted(img2,0.7,img1,0.3,0)
images = [img1,img2,output1,output2,output3]
plt.figure(figsize=(15,12))
for i in range(0,len(images)):
  plt.subplot(1,5,i+1)
  plt.axis("off")
  plt.imshow(images[i])
plt.show()
```

```python
img = cv2.imread("/content/monalisa.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gau_filter = cv2.GaussianBlur(img,(7,7),0)
images = [img,gau_filter]
for i in range(0,len(images)):
  plt.subplot(1,2,i+1)
  plt.axis("off")
  plt.imshow(images[i])
plt.show()
```
